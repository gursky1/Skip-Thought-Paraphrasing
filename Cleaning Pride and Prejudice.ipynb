```python
# Loading text
f=open('PrideandPred.txt', encoding = 'utf8')
raw=f.read()
raw = raw.replace('\ufeff','').replace('\n',' ').replace('”','"').replace('“','"').replace('  ',' ').replace('_','').replace('?"', '? "').replace('!"', '! "').replace('."', '. "').replace('Oh!','Oh,')
for i in range(0,100):
    raw = raw.replace('Chapter '+str(i)+' ','')

# Prepping tokenizer
sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

# Setting exceptions to tokenization
extra_abbreviations = set(['Dr', 'Mr', 'Mrs', 'Prof', 'Ms'])
sentence_tokenizer._params.abbrev_types.update(extra_abbreviations)

# Tokenizing text
sentences = sentence_tokenizer.tokenize(raw)

# Rejoining
sentences = '\n'.join(sentences).lower()

# Getting rid of endnotes
sentences = sentences.split('end of the project gutenberg ebook')[0]

# Splitting back up
sentences = sentences.split('\n')

# Padding punctuation with whitespace and removing leading/trailing whitespaces
sentences = [re.sub('([.,!?;()"])', r' \1 ', x).strip() for x in sentences]
sentences = [re.sub('\s{2,}', ' ', x).strip() for x in sentences]
sentences = [x for x in sentences if len(x)>0 ]
sentences = '\n'.join(sentences)
sentences = sentences.replace('mr . ','mr. ').replace('ms . ','ms. ').replace('mrs . ', 'mrs. ').replace('--',' : ')

# How many sentences in our corpus?
print(len(sentences.split('\n')))

# Removing leading and trailing whitespaces
print(*sentences.split('\n'), sep = '\n\n')
```

```python
# Loading text
f=open('MobyDick.txt', encoding = 'utf8')
raw=f.read()

# Dropping lines that have CHAPTER
raw = '\n'.join([x for x in raw.split('\n') if not 'CHAPTER' in x])

# Getting rid of characters
raw = raw.replace('\ufeff','').replace('\n',' ').replace('-',',').replace('”','"').replace('“','"').replace('  ',' ').replace('_','').replace('?"', '? "').replace('!"', '! "').replace('."', '. "').replace('Oh!','Oh,').replace('’',"'")

# Prepping tokenizer
sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

# Setting exceptions to tokenization
extra_abbreviations = set(['Dr', 'Mr', 'Mrs', 'Prof', 'Ms'])
sentence_tokenizer._params.abbrev_types.update(extra_abbreviations)

# Tokenizing text
sentences = sentence_tokenizer.tokenize(raw)

# Rejoining
sentences = '\n'.join(sentences).lower()

# Getting rid of endnotes
sentences = sentences.split('end of the project gutenberg ebook')[0]

# Splitting back up
sentences = sentences.split('\n')

# Padding punctuation with whitespace and removing leading/trailing whitespaces
sentences = [re.sub('([.,!?;()"])', r' \1 ', x).strip() for x in sentences]
sentences = [re.sub('\s{2,}', ' ', x).strip() for x in sentences]
sentences = [x for x in sentences if len(x)>0 ]
sentences = '\n'.join(sentences)
sentences = sentences.replace('mr . ','mr. ').replace('ms . ','ms. ').replace('mrs . ', 'mrs. ').replace('--',' : ')

# How many sentences in our corpus?
print(len(sentences.split('\n')))

# Removing leading and trailing whitespaces
print(*sentences.split('\n'), sep = '\n\n')
```
